```python
rows, cols = (ann.shape[0], ann.shape[1]) 
mask = [[0]*cols]*rows 
for i in range(ann.shape[0]):
    for j in range(ann.shape[1]):
        if ann[i][j][0]==4 and ann[i][j][1]==208 and ann[i][j][2]==5:
            mask[i][j]=1
```

```python
for i in range(0,5000,100):
    print(mask[12500][i])
```

```python
np.save('mask.npy', mask)
```
